```python
#Solution on Q1:
import numpy as np
np.random.seed(100)
arr = np.random.randint(1,11,size=(6, 10))
print('Original Array :' , arr, sep='\n')
# **Compute unique values and the row wise counts**

def row_wise_counts(arr2d):
    num_counts_array = [np.unique(row, return_counts=True) for row in arr2d]

    return([[int(b[a==i]) if i in a else 0
             for i in np.unique(arr2d)]
            for a, b in num_counts_array])

print('Unique values:  ',np.arange(1,11))
print('Row-wise count in Array for all unique values :' )
row_wise_counts(arr)
```
